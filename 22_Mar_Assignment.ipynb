Q1. To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you can use the `pearsonr` function from the `scipy.stats` module.

```python
from scipy.stats import pearsonr

# Sample data
studying_time = [5, 7, 6, 4, 8]  # hours
exam_scores = [80, 85, 75, 70, 90]

# Calculate Pearson correlation coefficient
corr_coef, p_value = pearsonr(studying_time, exam_scores)

print("Pearson correlation coefficient:", corr_coef)
```

Interpretation:
The Pearson correlation coefficient ranges from -1 to 1. A coefficient close to 1 indicates a strong positive linear relationship, meaning as studying time increases, exam scores tend to increase. A coefficient close to -1 indicates a strong negative linear relationship, meaning as studying time increases, exam scores tend to decrease. A coefficient close to 0 indicates no linear relationship between the variables.

Q2. To calculate Spearman's rank correlation between the amount of sleep individuals get each night and their overall job satisfaction level, you can use the `spearmanr` function from the `scipy.stats` module.

```python
from scipy.stats import spearmanr

# Sample data
sleep_hours = [7, 8, 6, 9, 7]  # hours
job_satisfaction = [8, 7, 6, 9, 8]

# Calculate Spearman's rank correlation
corr_coef, p_value = spearmanr(sleep_hours, job_satisfaction)

print("Spearman's rank correlation:", corr_coef)
```

Interpretation:
Spearman's rank correlation coefficient ranges from -1 to 1. A coefficient close to 1 indicates a strong positive monotonic relationship, meaning as one variable increases, the other tends to increase. A coefficient close to -1 indicates a strong negative monotonic relationship, meaning as one variable increases, the other tends to decrease. A coefficient close to 0 indicates no monotonic relationship between the variables.

Q4. To calculate the Pearson correlation coefficient between the number of hours individuals spend watching television per day and their level of physical activity, you can use the `pearsonr` function from the `scipy.stats` module similar to Q1.

```python
from scipy.stats import pearsonr

# Sample data
tv_hours = [3, 4, 2, 5, 1]  # hours
physical_activity = [2, 3, 1, 4, 1]  # level of activity (1 = low, 2 = moderate, 3 = high)

# Calculate Pearson correlation coefficient
corr_coef, p_value = pearsonr(tv_hours, physical_activity)

print("Pearson correlation coefficient:", corr_coef)
```

Interpretation:
The Pearson correlation coefficient ranges from -1 to 1. A positive coefficient indicates a positive relationship, meaning as one variable increases, the other tends to increase. A negative coefficient indicates a negative relationship, meaning as one variable increases, the other tends to decrease. A coefficient close to 0 indicates no linear relationship between the variables.

Q6. To calculate the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week, you can use the `pearsonr` function from the `scipy.stats` module similar to Q1.

```python
from scipy.stats import pearsonr

# Sample data
sales_calls_per_day = [20, 25, 18, 22, 15]  # calls
sales_per_week = [3, 4, 2, 3, 2]  # sales

# Calculate Pearson correlation coefficient
corr_coef, p_value = pearsonr(sales_calls_per_day, sales_per_week)

print("Pearson correlation coefficient:", corr_coef)
```

Interpretation:
The Pearson correlation coefficient will indicate the strength and direction of the linear relationship between the number of sales calls made per day and the number of sales made per week. A positive coefficient indicates a positive relationship, meaning more sales calls tend to result in more sales.
